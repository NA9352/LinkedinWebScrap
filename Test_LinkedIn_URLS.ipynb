```python
for url in df.loc[:, 'URLs']:
    sleep(1)
    driver.get(url)
    try:
        driver.find_element(By.CLASS_NAME, 'text-heading-xlarge')
        
        try:
            bio['Name'].append(driver.find_element(By.CLASS_NAME, 'text-heading-xlarge').text)
        except:
            bio['Name'].append('Not_Available')
        
        try:    
            bio['Headline'].append(driver.find_element(By.CLASS_NAME, 'text-body-medium.break-words').text)
        except:
            bio['Headline'].append('Not_Available')
        
        try:
            bio['Address'].append(driver.find_element(By.CSS_SELECTOR, 'span.text-body-small.inline.t-black--light.break-words').text)
        except:
            bio['Address'].append('Not_Available')
        
        #***************
        
        sleep(2)
        sections = driver.find_elements(By.CLASS_NAME, 'artdeco-card.ember-view.relative.break-words.pb3.mt2')
        #print(sections)
        sleep(2)
        for section in sections[:]:
            #print('n')
            if section.find_element(By.CLASS_NAME, 'pvs-header__container').text.split('\n')[0] == 'Experience':
            #   print('o')
                f = section.find_element(By.TAG_NAME, 'li')
                posi = f.find_elements(By.CLASS_NAME, 'visually-hidden')
                try:
                #   print('m')
                    bio['Position'].append(posi[0].text)
                except:
                    bio['Position'].append('None')
                    
                try:    
                    bio['Company_name'].append(posi[1].text)
                except:
                    bio['Company_name'].append('None')
                    
                try:
                    bio['Year'].append(posi[2].text)
                except:
                    bio['Year'].append('None')
                    
                try:
                    bio['Company_location'].append(posi[3].text)
                except:
                    bio['Company_location'].append('None')
                    
            if section.find_element(By.CLASS_NAME, 'pvs-header__container').text.split('\n')[0] == 'Education':
                list_tag = section.find_elements(By.TAG_NAME, 'li')
                for inde, a in enumerate(list_tag):
                    if inde > 1:
                        break
                    else:
                        try:
                            posi = a.find_elements(By.CLASS_NAME, 'visually-hidden')
                            bio[f'College_name_{inde+1}'].append(posi[0].text)
                            #bio[f'Degree_name_{inde+1}'].append(posi[1].text)
                            #bio[f'Passing_year_{inde+1}'].append(posi[2].text)
                        except:
                            bio[f'College_name_{inde+1}'].append('None')
                
                
        #*******  Contact Information **********#
        sleep(2)
        driver.get(url + 'overlay/contact-info/')

        info = BeautifulSoup(driver.page_source, 'lxml')
        details = info.findAll('section',attrs = {'class':'pv-contact-info__contact-type'})
        try:
            websites = details[1].findAll('a')
            for website in websites:
                website = website['href']
        except:
            website = 'None'
        bio['Website'].append(website)
            
        try:
            phone = details[2].find('span').text
        except:
            phone = 'None'
        bio['Phone'].append(phone)
        
        try:
            email = str(details[3].find('a').text).strip()
        except:
            email = 'None'
        bio['Email'].append(email)
    except:
        for item in list(bio.keys()):
            bio[item].append('Link doesnot work')
        
driver.close()


```

```python
OP = pd.DataFrame.from_dict(bio, orient='index').transpose()
OP = OP.transpose()
OP['URL'] = df['URLs'] 
OP.to_csv('Output.csv')
```
